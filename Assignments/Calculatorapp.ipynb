```python
input1 = int(input("Enter Value 1 : "))
input2 = int(input("Enter Value 2 : "))
operand = input("Enter your operand")

if operand == "+":
    print(input1 + input2)

elif operand == "-":
    print(input1 - input2)

elif operand == "/":
    print(input1 / input2)

elif operand == "x":
    print("The Multiplied value is : ", input1 * input2)

else:
    print("Please insert correct operator")
```
